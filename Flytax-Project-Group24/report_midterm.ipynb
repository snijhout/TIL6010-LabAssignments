# Introduction
The flight tax will be increased with €2,62 for departing passengers from the Netherlands as of January 1, 2024 (Welk Bedrag Betaal Ik Aan Vliegbelasting?, z.d.). 
After this increase, the new rate including indexation and increase comes to € 29,05 per departing passenger (Belastingdienst, 2024). This study will examine the impact of this increase on the number of departing flights and passengers. Finally, the influence of the air passenger tax on CO2 emissions will be examined.

# Research Objective 
## How does the implementation of the fly tax affect CO2 emissions from commercial flights in the Netherlands?

##### Sub-Questions

#### 1. How has the flight tax influenced the number of flights to and from the Netherlands?
#### 2. How has the flight tax impacted passenger demand for flights to and from the Netherlands and does this impact differ for intercontinental and european flights?

#### 3. To what extent will the flight tax contribute to reducing CO2 emissions for different flights to and from the Netherlands?
