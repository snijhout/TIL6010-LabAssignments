# Introduction
To encourage people to make more sustainable choices, the government has implemented a flight tax on all flights departing from Dutch airports. The aim is to encourage travelers to consider alternative, less polluting options, such as trains. 

In April 2024, this tax will increase by €2.62 for departing passengers from Dutch airports. Following this increase, the new flight tax will be €29.05 per departing passenger (belastingdienst, 2024). With the flight tax the Dutch government aims to reduce carbon emissions. 

This study aims to assess whether the flight tax truly affects people’s flight behavior by examining its impact on passenger demand. It also investigates the tax's effect on the number of flights departing from the Netherlands to determine whether the tax effectively reduces the number of flights. Finally, to provide a comprehensive analysis, the study explores the influence of the air passenger tax on overall CO2 emissions.

# 1. Analyses on flight tax influence on the number of flights
To evaluate the impact of the recent flight tax on the number of flights in the Netherlands, we first analyze flight data from 2000 to 2024, focusing on the second quarter of each year. We use second-quarter data because the available information on the tax impact only spans from April to August 2024. The graph below illustrates how flight numbers to and from the Netherlands have evolved over the past 24 years.

As shown in the graph, there appears to be no noticeable effect of the tax on aircraft movements. In fact, the increase from 2023 to 2024 is even steeper than that from 2022 to 2023. The absence of a visible impact may be due to the limited time frame available to observe any effects. In the future, however, we do expect changes as flight patterns continue to recover from the COVID-19 pandemic. Initially, we assumed that the effects of COVID-19 on air traffic had already stabilized, but this may not be entirely the case.

We also analyzed the annual growth in flight numbers over the past 24 years, as shown in the bar chart below. COVID-19, as seen in the previous graph, had a significant impact from 2020 to 2022, so these years are excluded from our growth comparison. To establish a baseline, we calculated the average annual growth from 2000 to 2020. This average was then compared with the growth observed from 2023 to 2024, following the implementation of the flight tax.

The average growth rate from 2000 to 2020 was 0.91%, while the increase from 2023 to 2024 was 4.66%. From these results can be concluded that the growth between 2023 and 2024 is even more than the growth before Covid and not even lower because of the tax. 

# 2. Analyses on flight tax influence on passenger demand and difference between intercontinental and european flights

This section will examine the number of passengers affected by the flight tax, building on the previous analysis of flight changes. It will differentiate between European and intercontinental flights. While the flight tax amount is the same for both categories, its impact can vary significantly. The graph below illustrates the number of passengers departing from and arriving in the Netherlands for both European and intercontinental destinations.

As depicted in the graph, there is no decline in passenger numbers for either category. However, unlike in Subsection 1, the number of intercontinental flights has not increased from 2023 to 2024. This stagnation is unlikely to be attributed to the flight tax; one would expect the tax to have a more pronounced effect on intercontinental flights if it were the sole factor influencing demand. A tax increase of €2.62 is likely to impact cheaper flights more significantly than those with higher fares.

This study also examines the annual growth of passengers from 2000 to 2020 in relation to the increase in passenger numbers between 2023 and 2024, the year the flight tax was introduced.

Between 2000 and 2020, the average annual growth rate for European passengers was 4.4%. In contrast, the growth rate for European passengers between 2023 and 2024 is projected to be 7.65%.

For intercontinental passengers, the average annual growth rate from 2000 to 2020 was 3.0%, while the growth rate from 2023 to 2024 is only expected to be 0.90%.
