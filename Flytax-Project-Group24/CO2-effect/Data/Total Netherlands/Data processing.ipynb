```python
import pandas as pd

# Path to the CSV file
file_path = r'C:\Users\omniq\Documents\GitHub\TIL6022-LabAssignments\Flytax-Project-Group24\CO2-effect\Data\fromNETHERLANDStin00190__custom_13279142_linear.csv'

# Load the CSV file
df = pd.read_csv(file_path)

# Create a dataframe with 'c_dest' (destination), 'TIME_PERIOD' (year), and 'OBS_VALUE' (second last column)
filtered_df = df[['c_dest', 'TIME_PERIOD', 'OBS_VALUE']]

# Group by 'TIME_PERIOD' and sum the 'OBS_VALUE' for each year
grouped_sum_df = filtered_df.groupby('TIME_PERIOD').agg({'OBS_VALUE': 'sum'}).reset_index()

# Display the total values summed by year
print("Summed OBS_VALUE per year:")
print(grouped_sum_df)

# Display the results per year with original data
grouped_df = filtered_df.groupby('TIME_PERIOD')

for year, data in grouped_df:
    print(f"\nYear: {year}")
    print(data)

```

```python
import pandas as pd

# File paths to the CSV files
file_path_from = r'C:\Users\omniq\Documents\GitHub\TIL6022-LabAssignments\Flytax-Project-Group24\CO2-effect\Data\Total Netherlands\fromNETHERLANDStin00190__custom_13279142_linear.csv'
file_path_to = r'C:\Users\omniq\Documents\GitHub\TIL6022-LabAssignments\Flytax-Project-Group24\CO2-effect\Data\Total Netherlands\toNETHERLANDStin00190__custom_13278956_linear.csv'

# Load both CSV files
df_from = pd.read_csv(file_path_from)
df_to = pd.read_csv(file_path_to)

# Create filtered dataframes with the relevant columns
filtered_df_from = df_from[['c_dest', 'TIME_PERIOD', 'OBS_VALUE']].rename(columns={'c_dest': 'Location', 'OBS_VALUE': 'Passengers From'})
filtered_df_to = df_to[['geo', 'TIME_PERIOD', 'OBS_VALUE']].rename(columns={'geo': 'Location', 'OBS_VALUE': 'Passengers To'})

# Group by 'TIME_PERIOD' and 'Location' for both datasets and sum the 'OBS_VALUE'
grouped_sum_df_from = filtered_df_from.groupby(['TIME_PERIOD', 'Location']).agg({'Passengers From': 'sum'}).reset_index()
grouped_sum_df_to = filtered_df_to.groupby(['TIME_PERIOD', 'Location']).agg({'Passengers To': 'sum'}).reset_index()

# Merge the two datasets based on 'TIME_PERIOD' and 'Location'
merged_df = pd.merge(grouped_sum_df_from, grouped_sum_df_to, on=['TIME_PERIOD', 'Location'], how='outer')

# Fill NaN values with 0 for missing data in either dataset
merged_df['Passengers From'] = merged_df['Passengers From'].fillna(0)
merged_df['Passengers To'] = merged_df['Passengers To'].fillna(0)

# Calculate the total passengers by summing the two columns
merged_df['Total Passengers'] = merged_df['Passengers From'] + merged_df['Passengers To']

# Display the merged dataframe showing passengers per year and location
print("Summed passengers per year and location (From and To Netherlands):")
print(merged_df)

# Filter the merged dataset for the year 2022
merged_df_2022 = merged_df[merged_df['TIME_PERIOD'] == 2022]

# Display the filtered dataframe for 2022
print("Passengers for the year 2022:")
print(merged_df_2022)


```
